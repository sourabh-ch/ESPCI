```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# B-I Fit

file_bi = 'B(I).txt'
file_bi_2 = 'B(I)-2.txt'

def data_bi(file):
    data = np.loadtxt(file, delimiter = '	')
    I = data[0:, 0]
    B = data[0:, 1]
    return I, B

i_1, b_1 = data_bi(file_bi)
i_2, b_2 = data_bi(file_bi_2)

m, b = np.polyfit(i_2, b_2, 1)

plt.figure()
plt.figure(figsize=(10,5))
plt.plot(i_2,b_2,'o')
plt.plot(i_2, m*i_2 + b)
plt.grid(color = 'grey', linestyle = '--', linewidth = 1)
plt.xlabel('Current I (A)')
plt.ylabel('Magnetic Field (T)')
```
